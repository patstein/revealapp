```python
def build_corpus(data):
    """Creates a list of lists containing words from each sentence
    
    Args:
        data: list or dataframe containing text in each cell.
    Returns: 
        A list of lists where the sublists are sentences
                and the sublist items are words."""
        
    corpus = []
    for content in data:
        corpus_temp = nltk.word_tokenize(content)
        corpus.append(corpus_temp)
    return corpus

corpus = build_corpus(pd.concat([data['sentence1'], data['sentence2']]))
```

```python
def avg_sentence_vector(sentence, model):
    #function to average all words vectors in a given paragraph/sentence
    words = word_tokenize(sentence)
    sentenceVec = model[words[0]]
    for i in range(1, len(words)):
        sentenceVec = sentenceVec + model[words[i]]
    return sentenceVec/len(words)
```
