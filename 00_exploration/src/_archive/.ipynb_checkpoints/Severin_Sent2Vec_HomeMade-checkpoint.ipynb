### Cleaning

```python
#Function for cleaning and stemming the data
def clean_sentence(val):
    "remove chars that are not letters or numbers, downcase, then remove stop words"
    regex = re.compile('([^\s\w]|_)+')
    sentence = regex.sub('', val).lower()
    sentence = re.sub("xxxx", "", sentence)
    sentence = re.sub("xxx", "", sentence)
    sentence = re.sub("xx", "", sentence)
    sentence = re.sub("\s\s+", " ", sentence)
       
    # stemming of words (seems not to affect accuracy, but should make things faster
   # porter = PorterStemmer()
   # words = word_tokenize(sentence)
   # sentence = " ".join([porter.stem(word) for word in words])
      
    sentence = sentence.split(" ")
    for word in list(sentence):
        if word in STOP_WORDS:
            sentence.remove(word)  
    sentence = " ".join(sentence)
    
    return sentence.strip() #remove spaces at the end and beginning


def clean_dataframe(data):
    "drop nans, then apply 'clean_sentence' function to question1 and 2"
    output_data = data[data['raw text'] == data['raw text']]  # removes nan since nan == nan -> False
    
    for col in ['raw text']:
        for row_idx, row in enumerate(output_data[col]):
            for idx, sentence in enumerate(output_data[col][row_idx]): #data[col][row] is a list of list strings, where each string is a sentence in this doc.
                output_data[col][row_idx][idx] = clean_sentence(sentence)
    
    return output_data

#splitting data into sentences
def tokenize_to_sentences(data):
    output_data = data.copy()
    for col in ['raw text']:
        output_data[col] = output_data[col].apply(nltk.sent_tokenize)
        
    return output_data
    
```

### Averaging by sentence

```python
#Calculate TF-IDF score for all words in the corpus
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=1, norm='l2', encoding='latin-1', ngram_range=(1,1), tokenizer=nltk.word_tokenize) #considering only 1-grams for now
features = tfidf.fit_transform(corpus_sentence).toarray()
#tfidf.vocabulary_ contains the mapping from the words to the score
```

```python
def get_sentence_vector(sentence=None, model=None, fitted_tfidf=None):
    '''
    This function is to average all words vectors in a given sentence.

    Parameters
    ----------
    sentence : string
            a string representing a sentence
    model : model
            a word embedding model (f.e. Word2Vec) featuring model['example_word']
    fitted_tfidf : fitted TfidfVectorizer
            pass the TF-IDF Vectorizer fitted to the dictionary
    '''
    words = word_tokenize(clean_sentence(sentence))
    
    #for debugging only: can be removed later
    if(len(words) == 0):
        print(sentence)
        print(clean_sentence(sentence))
        print(words)
        return None

    sent_len = len(words)
    total_weight = fitted_tfidf.vocabulary_[words[0]]
    sentenceVec = model.wv[words[0]] * fitted_tfidf.vocabulary_[words[0]]
    for i in range(1, sent_len):
        sentenceVec = sentenceVec + model.wv[words[i]] * fitted_tfidf.vocabulary_[words[i]]
        total_weight += fitted_tfidf.vocabulary_[words[i]]
                            
    return sentenceVec/total_weight
```

```python
def get_search_string(DocID = None, tag_idx = None, tag_length = None):
    return sample_text['raw text'][DocID][tag_idx:(tag_idx+tag_length)]
```

```python
def get_pivot_sentence(DocID = None, tag_idx = None, tag_length = None):
    '''
    This function finds the pivot sentence which is the sentence in which the selected tag is embedded in.

    Parameters
    ----------
    DocID : int
            index of document in document list
    tag_idx : int
            character index where tag starts
    tag_length : int
            length of the tag starting at tag_idx
    '''
    target_string = sample_text['raw text'][DocID]    
    sentences = nltk.sent_tokenize(target_string)

    pivot_sentence = sentences[0]
    global_idx = len(pivot_sentence) 
    sent_ctr = 0
    
    while(global_idx + len(pivot_sentence) < tag_idx):
        #move one sentence further
        sent_ctr += 1 #update counter
        global_idx += len(pivot_sentence) 
        pivot_sentence = sentences[sent_ctr]

    return pivot_sentence
```

```python
def get_most_similar (search_string = None, model = None, sentence_vectors = None, fitted_tfidf = None, n = 10):
    input_vector = get_sentence_vector(search_string, model, fitted_tfidf)
    sim = cosine_similarity(sentence_vectors, np.array(input_vector).reshape(1, -1)).T[0] #is a list of lists but the outer list contains only 1 element.
    toplist = pd.DataFrame({'sentences': all_sentences, 'similarity': sim}).sort_values(by = 'similarity', ascending = False)[0:n]
    return(toplist)
```
